```julia
include("xd.jl")
using .Xd
using Plots
```

```julia
function plotter(d,Ns,deltalatt,deltamc, sigmas,ylim1,xlim3)
    p1 = plot(xlabel="Points^1/"*string(d), ylabel="error")
    p1 = plot!(Ns,deltalatt,label="lattice")
    p1 = plot!(Ns,deltamc,label="MC")
    p1 = plot!(Ns,sigmas,label="SD of MC")
    p1 = plot!(ylim=ylim1)
    p2 = plot(Ns,deltalatt,ylim=(0,0.01),label="lattice",xlabel="Points^1/"*string(d), ylabel="error")
    Nsd= [N^d for N in Ns]
    p3= plot(xlabel="Points", ylabel="error")
    p3= plot!(Nsd,deltamc,label="MC")
    p3 = plot!(Nsd,sigmas,label="SD of MC")
    p3 = plot!(ylim=ylim1, xlim=xlim3)
    return p1,p2,p3
end
```

```julia
Ns1=[N for N in 1:100]
deltalatt1, deltamc1, sigmas1=Xd.precision(Ns1,1,100)
```

Ns1: 点数の次元乗根（１辺の刻み数）, deltalatt1: 台形公式の精度,  deltamc1: モンテカルロの精度, sigmas1: モンテカルロの標準偏差

```julia
d=1
p1,p2,p3=plotter(d,Ns1, deltalatt1, deltamc1, sigmas1,(-0.5,0.5),(0,100))
plot(p1,p2,p3,layout=(1,3),size=(1200,400),title=string(d)*" dimension")
```

lattice: 台形公式, 
MC: モンテカルロ, 
SD of MC: モンテカルロの標準偏差
